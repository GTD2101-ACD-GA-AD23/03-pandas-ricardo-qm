# Combinacion de datos

La combinaci√≥n de datos se refiere a la uni√≥n de dos DataFrames en base a una llave com√∫n. La combinaci√≥n de datos se puede realizar de tres formas:

* **Uno a uno:** Se combinan dos DataFrames en base a una llave com√∫n, donde la llave es √∫nica en ambos DataFrames.
* **Uno a varios:** Se combinan dos DataFrames en base a una llave com√∫n, donde la llave es √∫nica en el DataFrame de la izquierda y se repite en el DataFrame de la derecha.
* **Varios a varios:** Se combinan dos DataFrames en base a una llave com√∫n, donde la llave se repite en ambos DataFrames.

```python
canadian_youtube = pd.read_csv("data/CAvideos.csv")
british_youtube = pd.read_csv("data/GBvideos.csv")
```
